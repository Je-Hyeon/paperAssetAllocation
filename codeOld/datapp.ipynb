```python
import os 
import pandas as pd
```

```python
lst = [
    pd.read_pickle(f"data/raw/{i}") 
    for i in os.listdir("data/raw/")
]
cc = pd.concat(lst, axis=1).dropna()
cc.to_pickle("data/asset_class_data.pkl")
```

```python
lst = {
    i[:-4]: pd.read_pickle(f"data/assets/{i}") 
    for i in os.listdir("data/assets/")
}
cc = pd.concat(lst, axis=1).dropna(thresh=14)
cc.to_pickle('data/etf_cc.pkl')
```
